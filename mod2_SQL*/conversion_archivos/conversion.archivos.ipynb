```python
lista_datos = primer_elemento['0']

resultados = []

for datos in lista_datos:
    
    nombre = datos ['nombre']
    edad = datos ['edad']
    experiencia = datos ['experiencia']
    nivel = datos ['nivel']
    salario = datos ['salario']
    mascota = datos.get ('mascota', 'No tiene mascota')
    
    persona_info = {
        'nombre': nombre,
        'edad': edad,
        'experiencia': experiencia,
        'nivel': nivel,
        'salario': salario,
        'mascota': mascota
    }
    
    resultados.append (persona_info)
```
