str.extract(r'(\d+)\s+([a-zA-Z\s]+(?:\([A-Z]+\))?)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)'):

    (\d+): Captura un número (la posición).
    ([a-zA-Z\s]+(?:\([A-Z]+\))?): Captura el nombre del país y su código entre paréntesis.
    (\d+): Captura las medallas (oro, plata, bronce, total).
    Luego, extraemos todos estos grupos y los asignamos a nuevas columnas: Posición, País, Oro, Plata, Bronce, Total.

```python
# Usar str.split para separar la información
df_medallero[['Posición', 'País', 'Oro', 'Plata', 'Bronce', 'Total']] = df_medallero[0].str.extract(r'(\d+)\s+([a-zA-Z\s]+(?:\([A-Z]+\))?)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)')
```
